```python
import requests
from requests_html import HTML
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
options = Options()
options.add_argument("--headless") # for selenium to work without opening a web browser
driver = webdriver.Chrome(options = options)
```

```python

url = "https://www.amazon.in/Samsung-Galaxy-Storage-Additional-Exchange/dp/B089MQ7C7V/ref=sr_1_4?dchild=1&keywords=note+20&qid=1599050597&sr=8-4"
```

```python
title_lookup = "#titleSection"
price_lookup = "#priceblock_ourprice"
```

```python
driver.get(url)
body_el = driver.find_element_by_css_selector('body')
html_str = body_el.get_attribute("innerHTML")
print(html_str)
```

```python
html_obj = HTML(html = html_str)
```

```python
product_title = html_obj.find(title_lookup, first = True).text
```

```python
product_price = html_obj.find(price_lookup,first = True).text
```

```python
print(product_title)
print(product_price)
```
